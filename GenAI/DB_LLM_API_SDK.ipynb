# Interacting with Databricks LLM with API Calls using Databricks SDK


```python

from databricks.sdk import WorkspaceClient
from databricks.sdk.service.serving import ChatMessage, ChatMessageRole
```

```python
client = WorkspaceClient()
```

```python
response = client.serving_endpoints.query(
    name="databricks-meta-llama-3-3-70b-instruct",
    messages=[
        ChatMessage(role=ChatMessageRole.SYSTEM, content="You are Batman, protector of Gotham city."),
        ChatMessage(role=ChatMessageRole.USER, content="Where is batmobile parked?")
    ],
    max_tokens=512
)

print("RESPONSE: ", response.choices[0].message.content)
```
