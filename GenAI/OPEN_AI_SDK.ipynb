```python
from openai import OpenAI
import creds
```

```python
client = OpenAI(
    api_key=creds.DATABRICKS_DEVELOPER_API_TOKEN,
    base_url=f"{creds.DATABRICKS_WORKSPACE_HOSTNAME}/serving-endpoints"
)
```

```python
completion = client.chat.completions.create(
    model="databricks-llama-4-maverick",
    messages=[
        {
            "role": "system",
            "content": [
                {
                    "type": "text", "text": "You are a Batman, the protector of Gotham City."
                }
            ]
        },
        {
            "role": "user",
            "content": [
                {
                    "type": "text", "text": "How's Gotham city doing today?"
                }
            ]
        }
    ]
)
print(completion.choices[0].message.content)
```

# Sending image as Chat Request

```python
import base64
```

```python
!pwd
```

```python
image_path = "/Workspace/Users/darsh2115@gmail.com/demo/GenAI/assets/interrogation_scene.png"
```

```python
with open(image_path, "rb") as image_file:
    image_data = base64.standard_b64encode(image_file.read()).decode("utf-8")
```

```python
completion = client.chat.completions.create(
    model="databricks-llama-4-maverick",
    messages=[
        {
            "role": "system",
            "content": [
                {
                    "type": "text", "text": "You are a Batman, the protector of Gotham City."
                }
            ]
        },
        {
            "role": "user",
            "content": [
                {
                    "type": "text", "text": "Describe this image"
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/png;base64,{image_data}"
                    }
                }
            ]
        }
    ]
)
print(completion.choices[0].message.content)
```
